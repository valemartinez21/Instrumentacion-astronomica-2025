## Imagenes primera noche 

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy import units as u 
from astropy.visualization import ZScaleInterval
from astropy.coordinates import SkyCoord
from matplotlib import colormaps
```

```python
NGC7293V = fits.open('NGC_7293_V_80.000secs_00000411.fit')
NGC7293B = fits.open('NGC_7293_B_80.000secs_00000414.fit')
NGC7009B = fits.open('NGC_7009_B_40.000secs_00000433.fit')
NGC7009V = fits.open('NGC_7009_V_40.000secs_00000434.fit')
```

```python
data93v = NGC7293V[0].data
data93b = NGC7293B[0].data
data09b = NGC7009B[0].data
data09v = NGC7009V[0].data
```

## Nebulo HÃ©lice

```python
# NGC7293
zscale=ZScaleInterval(contrast=0.25)
min_value,max_value=zscale.get_limits(data93v)
min_value,max_value=zscale.get_limits(data93b)

fig,ax=plt.subplots(1,2, figsize=(12,6))
ax[0].imshow(data93b,vmin=min_value,vmax=max_value,cmap='gray',origin='lower')
ax[0].set_title('Imagen NGC 7293 B ')
ax[1].imshow(data93v,vmin=min_value,vmax=max_value,cmap='gray',origin='lower')
ax[1].set_title('Imagen NGC7293 V')



plt.tight_layout()
plt.show()
```
